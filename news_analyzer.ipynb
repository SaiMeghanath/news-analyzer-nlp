```python
from transformers import pipeline

# Load AI Models
summarizer = pipeline("summarization")
sentiment_analyzer = pipeline("sentiment-analysis")

print("✅ AI models loaded successfully!")
```

```python
news_text = """Apple has announced the launch of its latest iPhone model,
featuring a brand-new design, upgraded camera system, and improved battery life.
The company claims that the new device will revolutionize the smartphone industry
with its advanced AI-powered features."""

# Summarize the News
summary = summarizer(news_text, max_length=50, min_length=10, do_sample=False)

print("📜 News Summary:", summary[0]['summary_text'])
```

```python
sentiment = sentiment_analyzer(news_text)

print("🧐 Sentiment Analysis:", sentiment[0])
```

```python
import requests
from bs4 import BeautifulSoup

# URL of a news article
url = "https://www.bbc.com/news/world-asia-68010955"  # Change this to any news URL

# Fetch the webpage
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract the main article text
paragraphs = soup.find_all("p")
news_text = " ".join([para.get_text() for para in paragraphs])

print("📰 Extracted News:\n", news_text[:500])  # Print first 500 characters
```

```python
summary = summarizer(news_text, max_length=100, min_length=30, do_sample=False)

print("📜 News Summary:", summary[0]['summary_text'])
```

```python
sentiment = sentiment_analyzer(news_text)

print("🧐 Sentiment Analysis:", sentiment[0])
```
